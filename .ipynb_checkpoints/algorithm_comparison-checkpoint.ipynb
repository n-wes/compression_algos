```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('log.csv')
df = df.sort_values('original_size')

# Add ratio column
df['run_length_ratio'] = df.apply(lambda row: row['run_length_size']/row['original_size'], axis=1)
df['huffman_ratio'] = df.apply(lambda row: row['huffman_size']/row['original_size'], axis=1)
df
```

```python
# Size Comparisons
```

```python
ax = df.iloc[[6]].plot.bar(x='file_name', y=df.columns[1:4], rot=0, title='Size comparisons', figsize=(15,10), fontsize=12)
ax.set_xlabel('File Name')
ax.set_ylabel('Size (bits)')
plt.show()
```
